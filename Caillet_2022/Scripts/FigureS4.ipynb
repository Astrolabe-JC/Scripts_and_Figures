```python
# modules import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import xarray as xr
import seaborn as sns
```

```python
# Open Data
#model
netCDF_NEMO_file = '/Users/jcaillet/Documents/nemo/AMUXL025/ANALYSES/COMPLEMENT/comp_Dutrieux_REF.nc'
model = xr.open_mfdataset(netCDF_NEMO_file).sortby('time')
#mesure
mesure_file = '/Users/jcaillet/Documents/nemo/DATA_OBSERVATIONS/Vertical_profils/mesure_AMUXL025_19582018.nc'
mesure = xr.open_mfdataset(mesure_file).sortby('time')
```

```python
#Dutrieux
#year : 1994, 2000, 2007, 2009, 2010, 2012
mesure['year']=mesure['time'].dt.year
model['year']=model['time'].dt.year

# area : mask = 1 in the PIG area
mesure['tmask'] = mesure['longitude'] - mesure['longitude']
mesure['tmask'] = xr.where((mesure.longitude<-100) & (mesure.longitude>-104.5) & (mesure.latitude<-74) & (mesure.latitude>-75.5), 1, 0)
mesure['tmask'] = xr.where((mesure.longitude<-104.5) & (mesure.longitude>-110.5) & (mesure.latitude<-73.25) & (mesure.latitude>-75.5), 1, mesure.tmask.values)

mesure['T_conservative_b'] = mesure['T_conservative'] - mesure['T_conservative']
mesure['T_conservative_b'] = xr.where(mesure['T_conservative']==0, np.nan, mesure['T_conservative'])
mesure['S_absolute_b'] = mesure['S_absolute'] - mesure['S_absolute']
mesure['S_absolute_b'] = xr.where(mesure['S_absolute']==0, np.nan, mesure['S_absolute'])

model['T_conservative_b'] = model['T_conservative'] - model['T_conservative']
model['T_conservative_b'] = xr.where(model['T_conservative']==0, np.nan, model['T_conservative'])
model['S_absolute_b'] = model['S_absolute'] - model['S_absolute']
model['S_absolute_b'] = xr.where(model['S_absolute']==0, np.nan, model['S_absolute'])

mesure_select_meanPI = mesure.where(mesure.tmask==1).groupby(mesure.time.dt.year).mean()
model_select_meanPI = model.where(mesure.tmask==1).groupby(model.time.dt.year).mean()
mesure_select_stdPI = mesure.where(mesure.tmask==1).groupby(mesure.time.dt.year).std()
model_select_stdPI = model.where(mesure.tmask==1).groupby(model.time.dt.year).std()
```

```python
# Jenkins
#year : 2000, 2006, 2007, 2009, 2011, 2012, 2014, 2016
mesure['year']=mesure['time'].dt.year
model['year']=model['time'].dt.year
# area : mask = 1 in the Dotson area
mesure['tmask'] = mesure['longitude'] - mesure['longitude']
mesure['tmask'] = xr.where((mesure.longitude<-111.5) & (mesure.longitude>-113.6) & (mesure.latitude<-74.1) & (mesure.latitude>-74.25), 1, 0)

mesure['T_conservative_b'] = mesure['T_conservative'] - mesure['T_conservative']
mesure['T_conservative_b'] = xr.where(mesure['T_conservative']==0, np.nan, mesure['T_conservative'])
mesure['S_absolute_b'] = mesure['S_absolute'] - mesure['S_absolute']
mesure['S_absolute_b'] = xr.where(mesure['S_absolute']==0, np.nan, mesure['S_absolute'])

model['T_conservative_b'] = model['T_conservative'] - model['T_conservative']
model['T_conservative_b'] = xr.where(model['T_conservative']==0, np.nan, model['T_conservative'])
model['S_absolute_b'] = model['S_absolute'] - model['S_absolute']
model['S_absolute_b'] = xr.where(model['S_absolute']==0, np.nan, model['S_absolute'])

mesure_select_meanD = mesure.where(mesure.tmask==1).groupby(mesure.time.dt.year).mean()
model_select_meanD = model.where(mesure.tmask==1).groupby(model.time.dt.year).mean()
mesure_select_stdD = mesure.where(mesure.tmask==1).groupby(mesure.time.dt.year).std()
model_select_stdD = model.where(mesure.tmask==1).groupby(model.time.dt.year).std()
```
